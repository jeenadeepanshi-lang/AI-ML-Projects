```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.sex.unique()
```

```python
pd.get_dummies(df['sex']).head()
```

```python
sex = pd.get_dummies(df['sex'], drop_first = True)
sex.head()
```

```python
df.island.unique()
```

```python
pd.get_dummies(df['island']).head()
```

```python
island = pd.get_dummies(df['island'], drop_first = True)
island.head()
```

```python
new_data = pd.concat([df,island, sex], axis =1)
```

```python
new_data.head()
```

```python
new_data.drop(['sex', 'island'], axis =1, inplace = True);
```

```python
new_data.head()
```

```python
Y = new_data.species
Y.head()
```

```python
Y.unique()
```

```python
Y = Y.map({'Adelie':0, 'Chinstrap':1, 'Gentoo':2})
Y.head()
```

```python
new_data.drop('species', inplace =True, axis =1)
```

```python
new_data.head()
```

```python
X = new_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
print("X_train",X_train.shape )
print("X_test",X_test.shape)
print("Y_train",Y_train.shape )
print("Y_test",X_test.shape )
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 5, criterion = 'entropy', random_state =0)
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
```

```python
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```

```python
accuracy_score(Y_test, y_pred)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 5, criterion = 'gini', random_state =0)
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred)
```
